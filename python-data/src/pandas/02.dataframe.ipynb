## Data frame
DataFrame是矩形的数据表，它含有一组**有序且有命名的列**，每一列可以是不同的数据类型（数值、字符串、布尔值等）。DataFrame既有行索引也有列索引，可以看作由共用同一个索引的Series组成的字典。

```python
data = {"state": ["ohio", "ohio", 'ohio', 'dallas', 'dallas','dallas'], 
        "year": [2020,2021,2022,2022,2023,2024], 
        "pop": [1.5,1.6, 1.7, 2.3,2.4,2.6]
        }

frame = pd.DataFrame(data)

frame
```
