```python
import matplotlib.pyplot as plt
# 正态分布
x1 = np.random.normal(1.75,1,1_00_000_000)
plt.figure(figsize=(20,8),dpi=100)

plt.hist(x1, 1000)

plt.show()
```

```python
# 均匀分布
x2 = np.random.uniform(-1,1, 10000000)

plt.figure(figsize=(20,8), dpi=1000)
plt.hist(x2,1000)
plt.show()
```
