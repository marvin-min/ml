```python
import random
#0.准备数据
x= range(60)
y=[random.uniform(15,18) for i in x]
bj=[random.uniform(3,8) for i in x]
#1.创建画布
plt.figure(figsize=(20,8), dpi=100)
#显示网格
plt.grid(True,linestyle='--', alpha=0.5)
plt.plot(x,y,label='Shanghai')
plt.plot(x,bj, color='r', linestyle='--', label='Beijing')
x_labels = ["11:{}".format(i) for i in x]
y_ticks=range(40)
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("Temperature of XiAn at noon", fontsize=20)
plt.xticks(x[::5],x_labels[::5])
plt.yticks(y_ticks[::5])
#2.绘制图形
#要放在show的前面
plt.savefig("tst.jpg")
# 显示图例
plt.legend(loc='best')
#释放资源，放在show后面不会显示
plt.show()
```

### 多个坐标系绘图
plt.subplots(面向对象)

```python
# 0.准备数据
x = range(60)
shanghai = [random.uniform(15, 20) for i in x]
beijing =  [random.uniform(3, 12) for i in x]

# 1. 创建画布
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=100)

# 2.绘制图像
axes[0].plot(x,shanghai,label='ShangHai')
axes[1].plot(x,beijing,label='BeiJing', color="r", linestyle='--')

# 2.1 添加x,y周刻度
# 构造x，y轴刻度标签
x_ticks_label = ["11:{}".format(i) for i in x]
y_ticks = range(40)

#刻度显示
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks[::5])
axes[0].set_xticklabels(x_ticks_label[::5])


axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks[::5])
axes[1].set_xticklabels(x_ticks_label[::5])

# 2.2 添加网格显示
axes[0].grid(True, linestyle='--',alpha=0.5)
axes[1].grid(True, linestyle='--',alpha=0.5)

axes[0].set_xlabel("Time")
axes[0].set_ylabel("Temperature")
axes[0].set_title("Temperature of city", fontsize=20)

axes[1].set_xlabel("Time")
axes[1].set_ylabel("Temperature")
axes[1].set_title("Temperature of city", fontsize=20)

# 2.4保存图像
plt.savefig("test.png")

# 2.5 添加图例
axes[0].legend(loc=0)
axes[1].legend(loc=0)

# 3图像显示
plt.show()
```
