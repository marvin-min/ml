# 绘制多种类型图片

```python
import matplotlib.pyplot as plt
```

```python
#数据准备
x=[129,239,90,995,888]
y=[223,413,134,234,908]
```

```python
plt.figure(figsize=(20,8), dpi=100)
plt.scatter(x,y)
plt.show()
```

```python
# 绘制柱状图


movie_name =['abc','ddfe','adfadsf','adfasdfas','adf','df1','dfa','dd']
x=range(len(movie_name))
y=[7882,9001,10003,6900,600,8710,908,10900]

plt.figure(figsize=(20,8), dpi=100)

plt.bar(x,y,width=0.5, color=['b','c','y','m','g','b'])

plt.xticks(x, movie_name)
plt.grid(linestyle='--', alpha=0.5)
plt.title("movie income")
plt.show()
```
